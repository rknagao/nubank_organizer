```python
import os
import pandas as pd
import re
```

```python
# Levanta todos os dados na pasta clean_data
path_raw = r'C:\Users\Rafael\Documents\GitHub\nubank_organizer\raw_data'
list_f_raw = os.listdir(path_raw)
print(list_f_raw)
```

```python
# Abre o arquivo e salva as linhas em uma lista.
list_l_raw = []
last_f_raw = list_f_raw[-1]
print(last_f_raw)
with open(f'{path_raw}\{last_f_raw}', 'r', encoding='utf-8') as f:
    for l in f:
        list_l_raw.append(l)

list_l_raw
```

```python
# Remover recebimentos (valor negativo)
list_l_clean = [l for l in list_l_raw if '-R$' not in l]
```

```python
# Para cada linha, identificar o nome do produto.
list_name_raw = []
for l in list_l_clean:
# Encontra o índice do "R$"
    end = l.find("R$")

    # Extrai a substring entre os índices
    list_name_raw.append(l[6:end].strip())

list_name_raw
```

```python
# Remover subtextos desnecessários de produto.
regex = r" - Parcela \d+/\d+"
list_name_clean = []
for n in list_name_raw:
    name_clean = re.sub(regex, "", n)
    list_name_clean.append(name_clean)

list_name_clean
```

```python
# Para cada linha, identificar o valor do produto.
list_value_clean = []
for l in list_l_clean:
# Encontra o índice do "R$"
    end = l.find("R$")

    # Extrai a substring entre os índices
    list_value_clean.append(l[end+2:].strip())

list_value_clean
```

```python
# Juntar em um dataframe.
df = pd.DataFrame(list(zip(list_name_clean, list_value_clean)), columns=['Produto', 'Valor'])
df.head()
```

```python
# Correção de valor.
df['Valor'] = df['Valor'].str.replace('.', '')
df['Valor'] = df['Valor'].str.replace(',', '.').astype(float)
df.head()
```

```python
# Remover valor negativo: pagamento da referência anterior.
df = df.loc[df['Valor'] > 0].reset_index(drop=True)
df.head()
```

```python
# Agregar por produto.
df.groupby(['categoria', 'Produto']).agg({'Valor':'sum'})
```
